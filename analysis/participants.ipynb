```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame based on the information from the image
data = {
    'Sex': ['Male', 'Female'],
    'Count': [12, 3],
    'Average Age': [22, 21],
    
    'Ethnicity - Asian': [9, 1],
    'Ethnicity - White/Mixed': [3, 2],
    
    'Performed Exercise Before Yes': [12, 2],
    'Performed Exercise Before No': [0, 1],
    
    'Gym Consistency (2-3 times)': [5, 1],
    'Gym Consistency (3+ times)': [7, 0],
    'Gym Consistency (Never)': [0, 2]
}

df = pd.DataFrame(data)

# Display the data
print("Participant Demographics:")
print(df)

# Plot a bar chart of participant demographics
plt.figure(figsize=(10, 6))
plt.bar(df['Sex'], df['Count'])
plt.title('Participant Count by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```
